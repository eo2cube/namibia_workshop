## xarray indexing

```python
import datacube
dc = datacube.Datacube(app = 'xarray', config = '/home/datacube/.datacube.conf')
```

```python
dataset = dc.load(product = 's2_l2a_bavaria',
                  y= (5545645, 5545685),
                  x=(602955, 602975),
                  crs='EPSG:25832',
                  time=('2015-07-01', '2021-03-31'))
```

```python
dataset.red.isel(time=2)
```
